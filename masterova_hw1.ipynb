Преобразуйте датасет в разреженную матрицу scipy.sparse.csr_matrix, где значение x в позиции (i, j)
означает, что в документе i слово j встретилось x раз

```python
from functools import reduce
_dict = dict.fromkeys([*reduce(lambda a, x: a + x, data_tok)])
```

```python
__dict = dict([*zip(_dict.keys(),range(len(_dict)))])
__dict
```

```python
from collections import Counter
from scipy.sparse import csr_matrix
M = len(data_tok)
N = len(__dict)
dt = list()
indices = list()
indptr = list()
cur = 0
for line in data_tok:
    dt.extend(dict(Counter(line)).values())
    for word in dict(Counter(line)).keys():
        indices.append(__dict[word])
    indptr.extend([cur for i in range(len(dict(Counter(line)).keys()))])
    cur += 1

train = csr_matrix((dt, (indptr, indices)), shape=(M, N))
```

Реализуйте разбиение выборки для кросс-валидации на 3 фолдах. Разрешено использовать sklearn.cross_validation

```python
X_train, X_test, y_train, y_test = train_test_split(train, target,  test_size = 0.3, random_state = 11)

N_train = X_train.shape[0] 
N_test = X_test.shape[0] 
print (N_train, N_test)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)
print (err_train, err_test)
```
